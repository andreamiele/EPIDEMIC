```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
# Load the CSV file into a pandas DataFrame
path_ ='/content/ppo-fix.csv'
def plot(path_):
  name= path_.split("/")[-1].split(".")[0]

  print(name)
  csv_file_path = path_
  df = pd.read_csv(csv_file_path)

  # Define a function to apply EMA
  def apply_ema(series, alpha):
      return series.ewm(alpha=alpha, adjust=False).mean()

  # Define alphas
  alpha = 0.01
  alpha2 = 0.10

  # Apply EMA to the columns with different alphas
  df['counts/Susceptible'] = apply_ema(df['counts/Susceptible'], alpha=alpha)
  df['counts/Susceptible__MIN'] = apply_ema(df['counts/Susceptible__MIN'], alpha=alpha2)
  df['counts/Susceptible__MAX'] = apply_ema(df['counts/Susceptible__MAX'], alpha=alpha2)
  df['counts/Infected'] = apply_ema(df['counts/Infected'], alpha=alpha)
  df['counts/Infected__MIN'] = apply_ema(df['counts/Infected__MIN'], alpha=alpha2)
  df['counts/Infected__MAX'] = apply_ema(df['counts/Infected__MAX'], alpha=alpha2)
  df['counts/Recovered'] = apply_ema(df['counts/Recovered'], alpha=alpha)
  df['counts/Recovered__MIN'] = apply_ema(df['counts/Recovered__MIN'], alpha=alpha2)
  df['counts/Recovered__MAX'] = apply_ema(df['counts/Recovered__MAX'], alpha=alpha2)
  df['counts/Vaccinated'] = apply_ema(df['counts/Vaccinated'], alpha=alpha)
  df['counts/Vaccinated__MIN'] = apply_ema(df['counts/Vaccinated__MIN'], alpha=alpha2)
  df['counts/Vaccinated__MAX'] = apply_ema(df['counts/Vaccinated__MAX'], alpha=alpha2)


  # Plot the metrics with enhanced aesthetics and shaded areas
  plt.figure(figsize=(8, 8))

  # Susceptible
  plt.plot(df['global_step'], df['counts/Susceptible'], label='Susceptible', color="#4d6df3", linewidth=5)
  plt.fill_between(df['global_step'], df['counts/Susceptible__MIN'], df['counts/Susceptible__MAX'], color="#4d6df3", alpha=0.2)

  # Infected
  plt.plot(df['global_step'], df['counts/Infected'], label='Infected', color="#ed1c24", linewidth=5)
  plt.fill_between(df['global_step'], df['counts/Infected__MIN'], df['counts/Infected__MAX'], color="#ed1c24", alpha=0.2)

  # Recovered
  plt.plot(df['global_step'], df['counts/Recovered'], label='Recovered', color="#6f3198", linewidth=5)
  plt.fill_between(df['global_step'], df['counts/Recovered__MIN'], df['counts/Recovered__MAX'], color="#6f3198", alpha=0.2)

  # Vaccinated
  plt.plot(df['global_step'], df['counts/Vaccinated'], label='Vaccinated', color="#22b14c", linewidth=5)
  plt.fill_between(df['global_step'], df['counts/Vaccinated__MIN'], df['counts/Vaccinated__MAX'], color="#22b14c", alpha=0.2)

  # Add labels and title with improved font settings
  plt.xlabel('Global Step', fontsize=30)
  plt.ylabel('Population percentage', fontsize=30)
  plt.title('SIRV population', fontsize=30, fontweight='bold')

  # Customize the legend
  plt.legend(title='Population', title_fontsize='25', fontsize='20')

  # Add grid with custom settings
  plt.grid(True, linestyle="dotted", linewidth=0.5, color='gray', alpha=0.7)

  # Customize ticks
  plt.xticks(fontsize=30)
  plt.yticks(fontsize=30)
  # Increase the size of the 10e^6 at the end of the x-axis
  ax = plt.gca()
  ax.xaxis.set_major_formatter(ScalarFormatter(useMathText=True))
  plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
  ax.xaxis.get_offset_text().set_size(30)
  # Save the plot as a PDF file
  # Set x-axis limit
  plt.xlim(0, 5e6)
  pdf_file_path = name+'.pdf'
  plt.tight_layout()
  plt.savefig(pdf_file_path, format='pdf')

  print(f"Plot saved as {pdf_file_path}")
plot(path_)
```
